```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("web_pages.csv") # verileri okudum
```

```python
web_pages=data["Stats_Access_Link"].values # Stats_Acces_link kolonunun degerlerini değişkenime atadım.
```

```python
web_pages # değişkenimin ekran çıktısı
```

# Kırpma kısmı 

```python
sayac = 0 
for page in web_pages: 
    https = "https" in page # page linkinde https olup olmadığı kontrolü
    if https == True: # https ise
        page = page[13:] # page dizimin 13.indisindek itibaren sona kadar al ve page'i güncelle
    else:
        page = page[12:] # page dizimin 13.indisindek itibaren sona kadar al ve page'i güncelle
    page = page[:-6] # sondaki url tagını silme işlemi
    web_pages[sayac] = page # ana dizimin içine kırpılmış değerleri atadım.
    print(page)
    sayac += 1
```

```python
web_pages # ana dizimin çıktısı
```

```python
for index in data.index: # verimin içindeki satır sayısı ile döngü yaptım.
    print(index)
    data["Stats_Access_Link"].iloc[index] = web_pages[index] # dataframe'in Stats_Access_Link colomunun satırlarını güncelleme.
    
```
